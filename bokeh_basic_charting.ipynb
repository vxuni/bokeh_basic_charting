```python
import pandas as pd
eth_price = pd.read_html('https://coinmarketcap.com/currencies/ethereum/historical-data/', 'Market Cap', parse_dates=[0])[0]
eth_price.head()
```

```python
eth_price.rename({'Open*': 'Open', 'Close**': 'Close', 'Market Cap': 'Market_Cap'}, axis='columns', inplace=True)
```

```python
from bokeh.io import output_notebook, show

output_notebook()

import bokeh_basic_charting as bbc

p = bbc.make_candlesticks(eth_price)

show(p)  # open a browser
```

```python
p = bbc.make_candlesticks_cs(eth_price)
show(p)
```

```python
p.document.to_json()
```

```python
# make new figure; bokeh figures should not be reused like this, and deepcopy errors out.
p2 = bbc.make_candlesticks(eth_price)
p2 = bbc.add_volume_bars(eth_price, p2)
show(p2)
```

```python
pg = bbc.make_linked_candlesticks_and_volume(eth_price)
show(pg)
```
